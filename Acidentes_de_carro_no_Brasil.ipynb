# 5) Criando gráficos interativos:

```python
# Calcular a frequência de acidentes por dia da semana
acidentes_por_dia = df['dia_semana'].value_counts().reset_index()
acidentes_por_dia.columns = ['dia_semana', 'quantidade']

# Ordenar o DataFrame pela quantidade de acidentes em ordem decrescente
acidentes_por_dia = acidentes_por_dia.sort_values('quantidade', ascending=False)

# Criar o gráfico de barras com Plotly Express
fig = px.bar(acidentes_por_dia,
             x='dia_semana',
             y='quantidade',
             title='Quantidade de Acidentes por Dia da Semana (2017-2023) - Ordem Decrescente',
             labels={'dia_semana': 'Dia da Semana', 'quantidade': 'Quantidade de Acidentes'},
             template='plotly_white') # Você pode mudar o template se quiser

# Exibir o gráfico
fig.show()
```

```python

# Calcular a frequência de acidentes por estado (UF)
acidentes_por_estado = df['uf'].value_counts().reset_index()
acidentes_por_estado.columns = ['estado', 'quantidade']

# Ordenar o DataFrame pela quantidade de acidentes em ordem decrescente
acidentes_por_estado = acidentes_por_estado.sort_values('quantidade', ascending=False)

# Criar o gráfico de barras com Plotly Express
fig = px.bar(acidentes_por_estado,
             x='estado',
             y='quantidade',
             title='Quantidade de Acidentes por Estado (2017-2023) - Ordem Decrescente',
             labels={'estado': 'Estado (UF)', 'quantidade': 'Quantidade de Acidentes'},
             template='plotly_white') # Você pode mudar o template se quiser

# Exibir o gráfico
fig.show()
```

```python
# Calcular a frequência de acidentes por município
acidentes_por_municipio = df['municipio'].value_counts().reset_index()
acidentes_por_municipio.columns = ['municipio', 'quantidade']

# Selecionar os top 15 municípios
top_15_municipios = acidentes_por_municipio.head(15)

# Criar o gráfico de barras com Plotly Express para os top 15 municípios
fig = px.bar(top_15_municipios,
             x='municipio',
             y='quantidade',
             title='Top 15 Municípios com Mais Acidentes (2017-2023)',
             labels={'municipio': 'Município', 'quantidade': 'Quantidade de Acidentes'},
             template='plotly_white') # Você pode mudar o template se quiser

# Exibir o gráfico
fig.show()
```

```python
# Calcular a frequência dos tipos de acidente
tipos_acidente = df['tipo_acidente'].value_counts().reset_index()
tipos_acidente.columns = ['tipo_acidente', 'quantidade']

# Ordenar o DataFrame pela quantidade em ordem decrescente
tipos_acidente = tipos_acidente.sort_values('quantidade', ascending=False)

# Criar o gráfico de barras com Plotly Express
fig = px.bar(tipos_acidente,
             x='tipo_acidente',
             y='quantidade',
             title='Quantidade de Acidentes por Tipo (2017-2023) - Ordem Decrescente',
             labels={'tipo_acidente': 'Tipo de Acidente', 'quantidade': 'Quantidade de Acidentes'},
             template='plotly_white') # Você pode mudar o template se quiser

# Exibir o gráfico
fig.show()
```

```python
# Converter a coluna 'data_inversa' para datetime
df['data_inversa'] = pd.to_datetime(df['data_inversa'])

# Agrupar por data e contar o número de acidentes
acidentes_por_data = df['data_inversa'].value_counts().sort_index().reset_index()
acidentes_por_data.columns = ['data', 'quantidade']

# Criar o gráfico de linha com Plotly Express (sem a linha de tendência)
fig = px.line(acidentes_por_data,
              x='data',
              y='quantidade',
              title='Número de Acidentes ao Longo do Tempo (2017-2023)',
              labels={'data': 'Data', 'quantidade': 'Quantidade de Acidentes'},
              template='plotly_white')

# Exibir o gráfico
fig.show()
```

# Como a colisão traseira é o tipo de acidente mais comum, vamos verificar especificamente:

```python
# Filtrar o DataFrame para incluir apenas acidentes do tipo 'Colisão frontal'
colisao_frontal_df = df[df['tipo_acidente'] == 'Colisão frontal']

# Calcular a frequência das causas de acidente para colisões frontais
causas_colisao_frontal = colisao_frontal_df['causa_acidente'].value_counts().reset_index()
causas_colisao_frontal.columns = ['causa_acidente', 'quantidade']

# Ordenar o DataFrame pela quantidade em ordem decrescente e selecionar as top 5
top_5_causas_colisao_frontal = causas_colisao_frontal.sort_values('quantidade', ascending=False).head(5)

# Criar o gráfico de barras com Plotly Express usando a paleta Set2
fig = px.bar(top_5_causas_colisao_frontal,
             x='causa_acidente',
             y='quantidade',
             title='Top 5 Causas de Acidentes para Colisões Frontais (2017-2023)',
             labels={'causa_acidente': 'Causa do Acidente', 'quantidade': 'Quantidade'},
             template='plotly_white',
             color='causa_acidente', # Colorir as barras pela causa
             color_discrete_sequence=px.colors.qualitative.Set2) # Usar a paleta Set2

# Exibir o gráfico
fig.show()
```

```python
# Filtrar o DataFrame para incluir apenas acidentes do tipo 'Colisão frontal'
colisao_frontal_df = df[df['tipo_acidente'] == 'Colisão frontal']

# Calcular a frequência da classificação de acidente para colisões frontais
classificacao_colisao_frontal = colisao_frontal_df['classificacao_acidente'].value_counts().reset_index()
classificacao_colisao_frontal.columns = ['classificacao', 'quantidade']

# Criar o gráfico de rosca (donut chart)
fig = px.pie(classificacao_colisao_frontal,
             values='quantidade',
             names='classificacao',
             title='Classificação dos Acidentes para Colisões Frontais (2017-2023)',
             hole=0.4) # O valor 'hole' cria o furo no centro para fazer o donut chart

# Exibir o gráfico
fig.show()
```

```python
# Filtrar o DataFrame para incluir apenas acidentes do tipo 'Colisão frontal'
colisao_frontal_df = df[df['tipo_acidente'] == 'Colisão frontal']

# Calcular a frequência do dia da semana para colisões frontais
dia_semana_colisao_frontal = colisao_frontal_df['dia_semana'].value_counts().reset_index()
dia_semana_colisao_frontal.columns = ['dia_semana', 'quantidade']

# Definir a ordem correta dos dias da semana para o gráfico
ordem_dias = ['segunda-feira', 'terça-feira', 'quarta-feira', 'quinta-feira', 'sexta-feira', 'sábado', 'domingo']

# Reordenar o DataFrame para a ordem correta dos dias da semana no gráfico
dia_semana_colisao_frontal['dia_semana'] = pd.Categorical(dia_semana_colisao_frontal['dia_semana'], categories=ordem_dias, ordered=True)
dia_semana_colisao_frontal = dia_semana_colisao_frontal.sort_values('dia_semana')

# Criar o gráfico de barras com Plotly Express usando a paleta Set2
fig = px.bar(dia_semana_colisao_frontal,
             x='dia_semana',
             y='quantidade',
             title='Quantidade de Colisões Frontais por Dia da Semana (2017-2023)',
             labels={'dia_semana': 'Dia da Semana', 'quantidade': 'Quantidade de Acidentes'},
             template='plotly_white',
             color='dia_semana', # Colorir as barras pelo dia da semana
             color_discrete_sequence=px.colors.qualitative.Set2) # Usar a paleta Set2

# Remover a legenda
fig.update_layout(showlegend=False)

# Exibir o gráfico
fig.show()
```

```python
# Filtrar o DataFrame para incluir apenas acidentes do tipo 'Colisão frontal'
colisao_frontal_df = df[df['tipo_acidente'] == 'Colisão frontal']

# Calcular a frequência das condições meteorológicas para colisões frontais
condicoes_colisao_frontal = colisao_frontal_df['condicao_metereologica'].value_counts().reset_index()
condicoes_colisao_frontal.columns = ['condicao_metereologica', 'quantidade']

# Ordenar o DataFrame pela quantidade em ordem decrescente
condicoes_colisao_frontal = condicoes_colisao_frontal.sort_values('quantidade', ascending=False)

# Criar o gráfico de barras com Plotly Express e usar a paleta Set2
fig = px.bar(condicoes_colisao_frontal,
             x='condicao_metereologica',
             y='quantidade',
             title='Condições Meteorológicas para Colisões Frontais (2017-2023) - Ordem Decrescente',
             labels={'condicao_metereologica': 'Condição Meteorológica', 'quantidade': 'Quantidade de Acidentes'},
             template='plotly_white',
             color='condicao_metereologica', # Colorir as barras pela condição meteorológica
             color_discrete_sequence=px.colors.qualitative.Set2) # Usar a paleta Set2

# Remover a legenda
fig.update_layout(showlegend=False)

# Exibir o gráfico
fig.show()
```
